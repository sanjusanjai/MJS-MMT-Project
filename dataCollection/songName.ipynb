```python
topFiletoName={}
bottomFiletoName={}
relPath="./csvData/song_details_"
for i in range(2000,2022):
    topFiletoName[i]=[]
    bottomFiletoName[i]=[]
    fileName=relPath+str(i)+".csv"
    links=[]
    with open(fileName,'r') as f:
        temp=csv.reader(f)
        for row in temp:
            links.append({"link":row[-1],"name":row[0]})
    temp=[]
    for link in links:
        flag=0
        for x in temp:
            if x["link"]==link["link"]:
                flag=1
                break
        if flag==0:
            temp.append(link)
    links=temp
    
    topFileName="./mp3files/top/csv/"
    df=pd.read_csv(topFileName+str(i)+".csv")
    topData=df.to_dict('records')    
    for (j,dic) in enumerate(topData):
        for(k,songName) in enumerate(links):
            if dic["link"] in songName["link"]:
                topFiletoName[i].append({"fileName":dic["name"],"songName":songName["name"]})
    
    bottomFileName="./mp3files/bottom/csv/"
    df=pd.read_csv(bottomFileName+str(i)+".csv")
    bottomData=df.to_dict('records')
    for (j,dic) in enumerate(bottomData):
        for (k,songName) in enumerate(links):
            try:
                if dic["link"] in str(songName["link"]):
                    bottomFiletoName[i].append({"fileName":dic["name"],"songName":songName["name"]})
            except:
                pass
    
    df=pd.DataFrame(topFiletoName[i])
    df.to_csv("./mp3files/top/csv/fileNameToSong_"+str(i)+".csv",index=False,header=True)
    df=pd.DataFrame(bottomFiletoName[i])
    df.to_csv("./mp3files/bottom/csv/fileNameToSong_"+str(i)+".csv",index=False,header=True)
```
